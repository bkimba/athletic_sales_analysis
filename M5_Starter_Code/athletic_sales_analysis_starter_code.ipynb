```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
csv_path1 = "Resources/athletic_sales_2020.csv"
csv_path2 = "Resources/athletic_sales_2021.csv"

athletic_sales_2020_df = pd.read_csv(csv_path1)
athletic_sales_2021_df = pd.read_csv(csv_path2)
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_sales_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], ignore_index=True)
athletic_sales_df.head()
athletic_sales_df.tail()
```

```python
# Check if any values are null.
athletic_sales_df.isnull().sum()
```

```python
# Check the data type of each column
athletic_sales_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
athletic_sales_df["invoice_date"] = pd.to_datetime(athletic_sales_df["invoice_date"])
```

```python
# Confirm that the "invoice_date" data type has been changed.
athletic_sales_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
products_sold_df = athletic_sales_df.groupby(["region", "state", "city"]).agg({"product": "count"}).rename(columns={"product": "Total_Products_Sold"})



# Show the top 5 results.
products_sold_df.head()                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_athletic_sales_df = athletic_sales_df.pivot_table(index=["region", "state", "city"], values="product", aggfunc="count")

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_athletic_sales_df = pivot_athletic_sales_df.rename(columns={"product": "Total_Products_Sold"})

# Show the top 5 results.
pivot_athletic_sales_df.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales_df = athletic_sales_df.groupby(["region", "state", "city"]).agg({"total_sales": "sum"}).rename(columns={"total_sales": "Total Sales"})

# Show the top 5 results.
total_sales_df.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_as_product_df = athletic_sales_df.pivot_table(index=["region", "state", "city"], values="total_sales", aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_as_product_df = pivot_as_product_df.rename(columns={"total_sales": "Total Sales"})

# Show the top 5 results.
pivot_as_product_df.head()
```
